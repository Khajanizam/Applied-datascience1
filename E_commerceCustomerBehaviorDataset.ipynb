```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Distribution of Customer Age
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Distribution of Customer Age')
plt.xlabel('Customer Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Distribution of Annual Income
plt.figure(figsize=(8, 6))
sns.histplot(df['Annual Income'], bins=20, kde=True)
plt.title('Distribution of Annual Income')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Calculate Average Time on Site by Gender
average_time_on_site_by_gender = df.groupby('Gender')['Time on Site'].mean()

print(average_time_on_site_by_gender)

plt.figure(figsize=(8, 6))
sns.barplot(x=average_time_on_site_by_gender.index, y=average_time_on_site_by_gender.values)
plt.title('Average Time on Site by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Time on Site')
plt.show()
```

```python
# Calculate Average Time on Site by Location
average_time_on_site_by_location = df.groupby('Location')['Time on Site'].mean()

print(average_time_on_site_by_location)

plt.figure(figsize=(14, 6))
sns.barplot(x=average_time_on_site_by_location.index, y=average_time_on_site_by_location.values)
plt.title('Average Time on Site by Location (City Tier)')
plt.xlabel('City Tier')
plt.ylabel('Average Time on Site')
plt.show()
```
